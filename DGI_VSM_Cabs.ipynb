```python
#make a deep copy of cabs_transactions_customers
new_df = cabs_transactions_customers.copy(deep= True)
print(new_df.head())
```

```python
#make a list of categorical variables in cabs_transactions_customers
cat_cols = new_df.select_dtypes(include=['object']).columns.tolist()
print(cat_cols)
```

```python
#drop 'Date of Travel' from the list
cat_cols.remove('Date of Travel')
print(cat_cols)
```

```python
#dummy code categorical variables
new_df = pd.get_dummies(new_df, columns=cat_cols, drop_first=True)
print(new_df.head())
```

```python
#replace True with 1 and False with 0
new_df = new_df.replace({True: 1, False: 0})
print(new_df.head())
```

```python
#drop Transaction ID Date of Travel Customer ID
new_df = new_df.drop(['Transaction ID', 'Date of Travel', 'Customer ID'], axis=1)
print(new_df.head())
```

```python
#drop month, year
new_df = new_df.drop(['month', 'year'], axis=1)
print(new_df.head())
```

```python
#print columns
print(new_df.columns)
```

```python
#drop Income_Bin, Age_Bin
new_df = new_df.drop(['Income_Bin', 'Age_Bin'], axis=1)
print(new_df.head())
```

```python
#drop avg_profit_per_ride, Price Charged
new_df = new_df.drop(['avg_profit_per_ride', 'Price Charged'], axis=1)
print(new_df.head())
```

```python
#print corr matrix
corr_matrix = new_df.corr()
print(corr_matrix)
```

```python
#create interaction term between KM Travelled and Cost of Trip
new_df['KM_Cost_Interaction'] = new_df['KM Travelled'] * new_df['Cost of Trip']
print(new_df.head())
```

```python
#drop  KM Travelled and Cost of Trip
new_df = new_df.drop(['KM Travelled', 'Cost of Trip'], axis=1)
print(new_df.head())
```

```python
#create X_df by dropping profits
X = new_df.drop('profits', axis=1)
print(X.head())
```

```python
#create Y with profits column
y = new_df['profits']
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#fitting a linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#make prediction
y_pred = model.predict(X_val)
```

```python
#find accuracy
mse = mean_squared_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
